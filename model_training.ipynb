```python
import albumentations as A
from ultralytics import YOLO
from ultralytics.models.yolo.detect import DetectionTrainer
```

### Define Weather augmentations

```python
WEATHER_AUGS = [
    A.OneOf(
        [
            A.MotionBlur(blur_limit=7, p=1.0),
            A.MedianBlur(blur_limit=7, p=1.0),
            A.GaussianBlur(blur_limit=7, p=1.0),
        ],
        p=0.3,
    ),
    
    A.OneOf([
        A.GaussianBlur(blur_limit=(3, 7), p=1.0),
        A.ISONoise(color_shift=(0.01, 0.05), intensity=(0.1, 0.5), p=1.0),
    ], p=0.2),
    A.RandomBrightnessContrast(
        brightness_limit=0.2,
        contrast_limit=0.2,
        p=0.3,
    ),
    A.OneOf([
        A.RandomRain(drop_length=12, drop_width=1, blur_value=2, brightness_coefficient=0.85, p=1.0),
        A.RandomFog(fog_coef_range=(0.05, 0.15), alpha_coef=0.08, p=1.0)
    ], p=0.2),
    A.CLAHE(clip_limit=4.0, tile_grid_size=(8, 8), p=0.5)
]
```

### Custom Trainer

```python
model = YOLO("yolo26n.pt", task="detect")

model.train(
    data="/media/mkd/Manav Desai/Datasets/bdd_subset/data.yml", epochs=10, batch=8, augmentations=WEATHER_AUGS
)
```
