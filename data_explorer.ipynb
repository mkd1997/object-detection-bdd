```python
import json
```

```python
train_json_path: str = "/media/mkd/Manav Desai/Datasets/bdd_dataset/bdd100k_labels_release/bdd100k/labels/bdd100k_labels_images_train.json"
```

```python
with open(train_json_path, "r") as f:
    train_annotation_json = json.load(f)
```

```python
train_annotation_json[0]
```

```python
category_set = set()
for img_annotation in train_annotation_json:
    for label in img_annotation["labels"]:
        category_set.add(label["category"])

category_set
```

```python
weather_set = set()
for img_annotation in train_annotation_json:
    weather_set.add(img_annotation["attributes"]["weather"])

weather_set
```

```python
scene_set = set()
for img_annotation in train_annotation_json:
    scene_set.add(img_annotation["attributes"]["scene"])

scene_set
```

```python
tod_set = set()
for img_annotation in train_annotation_json:
    tod_set.add(img_annotation["attributes"]["timeofday"])

tod_set
```

```python
timestamp_set = set()
for img_annotation in train_annotation_json:
    timestamp_set.add(img_annotation["timestamp"])

timestamp_set
```

```python
attribute_types = set()
for img_annotation in train_annotation_json:
    for attribute in img_annotation["attributes"]:
        attribute_types.add(attribute)

attribute_types
```

```python
label_attributes = set()
for img_annotation in train_annotation_json:
    for label in img_annotation["labels"]:
        if (label["category"] not in ("drivable area", "lane")):
            for attribute in label["attributes"]:
                label_attributes.add(attribute)

label_attributes
```
